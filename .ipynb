```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 10, 100)
```

```python
siny = np.sin(x)
cosy = np.cos(x)
```

```python
plt.plot(x, siny, label="sin(x)")
plt.plot(x, cosy, label="cos(x)", color="red", linestyle="--")
plt.xlabel("x axis")
plt.ylabel("y value")
#plt.xlim(0, 15)
#plt.ylim(0, 1.5)
plt.axis([-1, 16, -1, 2])
plt.legend()
plt.title("Test Matplotlib")
plt.show()
```
