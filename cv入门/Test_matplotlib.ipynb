```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets
```

```python
x = np.linspace(0, 10, 100)
```

```python
siny = np.sin(x)
cosy = np.cos(x)
```

```python
plt.plot(x, siny, label="sin(x)")
plt.plot(x, cosy, label="cos(x)", color="red", linestyle="--")
plt.xlabel("x axis")
plt.ylabel("y value")
#plt.xlim(0, 15)
#plt.ylim(0, 1.5)
#plt.axis([-1, 16, -1, 2])
plt.legend()
plt.title("Test Matplotlib")
plt.show()
```

```python
iris = datasets.load_iris()
iris.keys()
```

```python
iris.data.shape
```

```python
X = iris.data[:, :2]
```

```python
plt.scatter(X[:,0], X[:,1])
plt.show()
```

```python
y = iris.target
plt.scatter(X[y==0,0], X[y==0,1], color="red", marker="o")
plt.scatter(X[y==1,0], X[y==1,1], color="blue", marker="+")
plt.scatter(X[y==2,0], X[y==2,1], color="green", marker="x")
plt.show()
```
